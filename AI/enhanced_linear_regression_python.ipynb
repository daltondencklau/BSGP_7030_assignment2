# Enhanced Linear Regression (Python)

This notebook reconstructs a univariate ordinary least squares model relating an outcome variable to a single continuous predictor. The workflow mirrors the companion CLI script: ingest tabular data, estimate coefficients, quantify goodness-of-fit via $R^2$, and visualize residuals of the linear approximation against the empirical scatter.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

DATA_PATH = "regression_data.csv"
X_COL = "YearsExperience"
Y_COL = "Salary"

data = pd.read_csv(DATA_PATH)
data.head()
```

```python
x = data[[X_COL]]
y = data[Y_COL]

model = LinearRegression()
model.fit(x, y)
y_hat = model.predict(x)
r2 = model.score(x, y)

print(f"Intercept (β₀): {model.intercept_:.4f}")
print(f"Slope (β₁):     {model.coef_[0]:.4f}")
print(f"R²:             {r2:.4f}")
```

```python
fig, ax = plt.subplots(figsize=(7.5, 5))
ax.scatter(x, y, color="#c0392b", alpha=0.85, label="Observations")
ax.plot(x, y_hat, color="#2980b9", linewidth=2, label="OLS fit")
ax.set_title(f"{Y_COL} as a function of {X_COL}")
ax.set_xlabel(X_COL)
ax.set_ylabel(Y_COL)
ax.legend(frameon=False)
fig.tight_layout()
plt.show()
```
