# Enhanced Linear Regression (R)

Parallel implementation of the Python analysis using base-R `lm()` for parameter estimation and **ggplot2** for visualization. The notebook emphasizes interpretability of the fitted line, residual uncertainty (via `geom_smooth` confidence band), and the classical summary diagnostics produced by `summary(model)`.

```r
data <- read.csv("regression_data.csv")
head(data)
```

```r
model <- lm(Salary ~ YearsExperience, data = data)
summary(model)
```

```r
if (!requireNamespace("ggplot2", quietly = TRUE)) {
  install.packages("ggplot2", repos = "https://cloud.r-project.org")
}
library(ggplot2)

ggplot(data, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "#c0392b", alpha = 0.85) +
  geom_smooth(method = "lm", se = TRUE, color = "#2980b9", fill = "#AED6F1") +
  ggtitle("Salary as a function of YearsExperience") +
  xlab("YearsExperience") +
  ylab("Salary") +
  theme_minimal(base_size = 12)
```
