```python
## importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('regression_data.csv')
```

```python
df.head()
```

### initial data exploration

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

### creating a regression model

```python
## creating a linear model to the salary and years of experience df
lin_reg1 = LinearRegression()

## fitting the model to the df
lin_reg1.fit(df[['YearsExperience']], df['Salary'])
```

```python
## plotting the regression line to examine salary as a function of years of experience
plt.plot(df["YearsExperience"], lin_reg1.predict(df[["YearsExperience"]]), color="blue")
plt.title("Salary v. Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

### model evaluation

```python
## evaluating the model using R-Squared
r_sq = lin_reg1.score(df[["YearsExperience"]], df["Salary"])
```
